```python
import os,geopandas,yaml
n = r'U:\NonPerennialStreamPOC\examples\conus1\input\parflow_conus1_domain_huc8s.gpkg'
sd = r'U:\NonPerennialStreamPOC\examples\conus1\subdomains'
iyml = r'U:\NonPerennialStreamPOC\examples\conus1\namelist.yaml'
with open(iyml, 'r') as f:
    nl = yaml.safe_load(f)
i = geopandas.read_file(n)
for index,row in i.iterrows():
    huc8id = str(row['huc8'])
    hucdir = os.path.join(sd,huc8id)
    if not os.path.isdir(hucdir): os.mkdir(hucdir)
    nl['domain_huc'] = huc8id
    with open(os.path.join(hucdir,'namelist.yaml'), 'w') as f:
        yaml.dump(nl, f, sort_keys=False)
    inputdir = os.path.join(sd,huc8id,'input')
    if not os.path.isdir(inputdir): os.mkdir(inputdir)
    fd = os.path.join(inputdir,'domain.gpkg')
    i.loc[[index]].to_file(fd,driver='GPKG')
```
